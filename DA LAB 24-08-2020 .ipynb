### CHECK FOR MISSING OR WRONG VALUES

```python
for i in df.columns:
    print(i,' = ',df[i].unique())
    print()
```

```python
df2=df.loc[df['ca']!='?']
df2=pd.to_numeric(df2['ca'].unique())
df['ca']=(pd.to_numeric(df['ca'].replace('?',df2.mean()))).astype('int64') #converted to int as per the other data
```

```python
df2=df.loc[df['thal']!='?']
df2=pd.to_numeric(df2['thal'].unique())
df['thal']=(pd.to_numeric(df['thal'].replace('?',df2.mean()))).astype('int64') #converted to int as per the other data
```
